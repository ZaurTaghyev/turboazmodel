```python
df=pd.read_csv('t2.csv')
df[['Serfiyyat','At gucu','Yanacaq']]=df['Mühərrik'].str.split('/',expand=True)
df.drop('Mühərrik',axis=1,inplace=True)
df['Serfiyyat']=df['Serfiyyat'].str.replace(' L','')
df['At gucu']=df['At gucu'].str.replace(' a.g.','')
df['Yürüş']=df['Yürüş'].str.replace(' km','')
df[['Vurug','renglenmek']]=df['Vəziyyəti'].str.split(',',expand=True)
df.drop('Vəziyyəti',axis=1,inplace=True)
money=[]
def func(x):
    if x.split(' ')[-1]=='USD':
       money.append(int(x.replace(' USD','').replace(' ',''))*1.7)
    elif x.split(' ')[-1]=='EUR':
        money.append(int(x.replace(' EUR','').replace(' ',''))*1.8)
    else:
       money.append(int(x.replace(' AZN','').replace(' ','')))
df['Qiyməti'].apply(func)
df['Qiyməti']=money
df.drop('Unnamed: 0',axis=1,inplace=True)
df.drop('Unnamed: 0.1',axis=1,inplace=True)
for i in range(len(df)):
    if df['Yürüş'][i]==str(df['Yürüş'][i]):
        df['Yürüş'][i]=df['Yürüş'][i].replace(' ','')
        df['Yürüş'][i]=int(df['Yürüş'][i])
    elif df['Yürüş'][i]=='0':
         df['Yürüş'][i]=int(0)
    elif df['Serfiyyat'][i]==str(df['Serfiyyat'][i]):
         df['Serfiyyat'][i]=float(df['Serfiyyat'][i])
    elif df['At gucu'][i]==str(df['At gucu'][i]):
        df['At gucu'][i]=int(df['At gucu'][i])
# def find_boundaries(df, variable, distance = 1.5):
#     IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

#     lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
#     upper_boundary = df[variable].quantile(0.75) + (IQR * distance)

#     return upper_boundary, lower_boundary
# B_upper_limit, B_lower_limit = find_boundaries(df, 'Buraxılış ili')
# B_upper_limit, B_lower_limit
# # df['Buraxılış ili']=np.where(df['Buraxılış ili']>B_upper_limit,B_upper_limit,np.where(df['Buraxılış ili']<B_lower_limit,B_lower_limit,df['Buraxılış ili']))
# # df
```

```python
x=df.drop('Qiyməti',axis=1)
y=df['Qiyməti']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler=scaler.fit(x.select_dtypes(exclude='object'))
sc = scaler.transform(x.select_dtypes(exclude='object')) 
sc=pd.DataFrame(sc,columns=x.select_dtypes(exclude='object').columns)
x[x.select_dtypes(exclude='object').columns]=sc
```

```python
for n in x.select_dtypes(include='object'):
  encode=x.groupby(n).size()/len(x)
  x[n]=x[n].map(encode)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf_mse=DecisionTreeRegressor(criterion='friedman_mse',max_depth=19,random_state=0)
clf_mse.fit(x_train,y_train)
y_pred=clf_mse.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import math
print(mean_squared_error(y_test, y_pred))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
